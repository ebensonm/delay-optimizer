# Get hyperparameter optimization results

Run get_results.sh first to update the hyperparameter optimization results.

```python
# Read params
params = pd.read_csv("../hyperparameter_data.csv")
```

```python
def parse(args):
    new_params = params.copy()
    for arg, val in args.items():
        new_params = new_params[new_params[arg] == val]
        
    return new_params

def extract_params(df, num_param_sets):
    if len(df) > num_param_sets:
        df = df[:num_param_sets]
    vals = df.loc[:,['max_learning_rate', 'min_learning_rate', 'step_size']].values
    for params in vals:
        print(params[0], ",", params[1], ",", params[2], sep="")
```

```python
# Drop columns
drop_cols = ["loss_name", "dim", "use_delays", "max_L", "constant_learning_rate", "learning_rate", 
             "min_val", "max_val", "num_delays", "maxiter", "optimizer_name", "tol", "max_evals",
             "symmetric_delays", "num_tests", "print_log", "vary_percent", "hyper_minimize", "clip_grad",
             "clip_val", "num_test_initials"]
```

```python
# 2d undelayed
args = {"loss_name":'Rosenbrock', "dim":2, "use_delays":False, "constant_learning_rate":False}
rosen2d_params = parse(args)
rosen2d_params = rosen2d_params.drop(columns=drop_cols)
rosen2d_params = rosen2d_params.sort_values('best_loss')
extract_params(rosen2d_params, 5)
```
